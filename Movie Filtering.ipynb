```python
movies= pd.read_csv("https://github.com/SavioSal/datasets/raw/master/movie_clean.csv",encoding= 'unicode_escape')
```

```python
movie_name= input("Enter a Movie name: ")
```

```python
movies=movies.fillna("")
```

```python
#determine the records for the movie name
selected_movie= movies[movies['title'].str.contains(movie_name)]
```

```python
selected_movie
```

```python
#get the geners of the selected movie
selected_movies_genre = selected_movie['genres']
```

```python
selected_movies_genre
```

```python
movies.values
```

```python
movie_index_counter = -1
movie_index_list = []
for ind_movie_genre in movies['genres']:
    movie_index_counter += 1
    for single_genre in selected_movies_genre:
        if single_genre in ind_movie_genre:
            movie_index_list.append(movie_index_counter)
            break
            
print ("Movie Index List: ", movie_index_list)
```

```python
movies.take(movie_index_list)
```
